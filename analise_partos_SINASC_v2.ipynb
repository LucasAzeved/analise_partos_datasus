```python
# Diretorio para download
p = Path().absolute() / 'Files_SINASC'
```

```python
# estados = ['AC','AL','AP','AM','BA','CE','DF','ES','GO',
#           'MA','MT','MS','MG','PA','PB','PR','PE','PI',
#           'RJ','RN','RS','RO','RR','SC','SP','SE','TO']


# Apenas estados do sul
estados = ['RS', 'SC', 'PR']

anos = [2021, 2020, 2019, 2018]

downloaded = sinasc.download(
    states=estados,
    years=anos,
    data_dir=p
    )

downloaded
```

```python
banco = {}

for f in p.glob('*'):
    banco[f.name] = str(f)
```

```python
banco
```

```python
parquets_to_dataframe?
```

```python
df = []
for f in banco.values():
    df.append(parquets_to_dataframe(str(f)))
    
df = pd.concat(df, axis=0).reset_index(drop=True)
```

```python
display(df.shape)
df.head()
```

```python
for c in df.columns:
    df[c] = df[c].replace('', np.nan)
```

```python
l = df.shape[0]

for i, c in enumerate(df.columns):
    v = df[c].notna().sum()
    print(f'{i:5} {c:15} {v:10} {(v/l)*100:6.2f} {df[c].dtype:}')
```

```python
cols = [
       'CODMUNNASC', 'IDADEMAE', 'ESTCIVMAE', 'GESTACAO', 'GRAVIDEZ', 'PARTO', 'CONSULTAS', 
       'DTNASCMAE', 'DTNASC', 'HORANASC', 'RACACOR', 'PESO', 'ESCMAE2010',  'RACACORMAE'
    ]

df = df[cols]
```

```python
cols_int = [
    'IDADEMAE',
    'PESO',
]

for c in cols_int:
    # if c in cols_int:
    if df[c].dtype == 'string':
        df[c] = df[c].fillna('-1')
        df[c] = df[c].astype(int)

```

```python
[c for c in df.columns if c not in cols_int]
```

```python
mappers = pd.read_excel('mappers.xlsx').astype(str)
```

```python
mappers['coluna'].value_counts()
```

```python
for c in mappers['coluna'].unique():
    if c in df.columns:
        dict_c = dict(mappers[mappers['coluna'] == c][['cod', 'descricao']].values)
        df[f'{c}'] = df[c].map(dict_c).fillna('Não Idenficado')
```

```python
for v in ('DTNASCMAE', 'DTNASC'):
    df[v] = df[v].fillna('01011900')
    df[v] = df[v].str.replace(' ', '').apply(lambda x: datetime(int((x[-4:])), int(x[2:4]), int(x[:2])))
    df.loc[df[v] == datetime(1900, 1, 1), v] = np.nan
```

```python
df['ANONASC'] = df['DTNASC'].apply(lambda x: x.year)
df['MESNASC'] = df['DTNASC'].apply(lambda x: x.month)
```

```python
df['HORANASC'] = df['HORANASC'].str.replace(' ', '')
df['HORANASC'] = df['HORANASC'].str.slice(0, 2)
```

```python
mun = pd.read_csv('Municipios_20220214.csv', dtype='O', encoding='ISO-8859-1', sep=';')
mun = mun[mun['UF_txt'].isin(('RS', 'PR', 'SC'))]
```

```python
mun_mapper = dict(
        zip(
            mun['Código Município Completo'].str.slice(0, -1).values, 
            mun['Nome_Município'].values
        )
    )
uf_mapper = dict(
        zip(
            mun['Código Município Completo'].str.slice(0, -1).values, 
            mun['UF_txt'].values
        )
    )
```

```python
df['NOMEMUNNASC'] = df['CODMUNNASC'].str.replace(' ', '').map(mun_mapper)
df['UFNASC'] = df['CODMUNNASC'].str.replace(' ', '').map(uf_mapper)
df = df[df['NOMEMUNNASC'].notna()]
```
